# 1. Growth Of Function
Time Complexity/Order of Growth defines the amount of time taken by any program with respect to the size of the input. Time Complexity specifies how the program would behave as the order of size of input is increased. So, Time Complexity is just a function of size of its input. We will go through some of basic and most common time complexities such as:

- *Constant Time Complexity O(1) : constant running time*
- *Linear Time Complexity O(n) : linear running time*
- *Logarithmic Time Complexity O(log n) : logarithmic running time*
- *Log-Linear Time Complexity O(n log n) : log-Â­linear running time*
- *Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)*
- *Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)*

## Constant time Complexity:O(1)
An algorithm is said to have constant time with order O (1) when it is not dependent on the input size n. Irrespective of the input size n, the runtime will always be the same. See example below

```python
# Load the Python libraries
import matplotlib.pyplot as plt
```

```python
# Prob_1

def add_two_numbers(a, b):
    return a + b

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    add_two_numbers(n, n + 1)
    time_values.append(1)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(1) Constant Time Complexity for Addition")
plt.show()
```

#Linear Time Complexity O(n) : linear running time
- *An algorithm is said to have a linear time complexity when the running time increases linearly with the length of the input. When the function involves checking all the values in input data, with this order O(n).*

```python
def linear_time_sum(numbers):
    total = 0
    for value in numbers:
        total += value
    return total

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    linear_time_sum(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Logarithmic Time Complexity O(log n) : logarithmic running time
- *An algorithm is said to have a logarithmic time complexity when it reduces the size of the input data in each step. This indicates that the number of operations is not the same as the input size. The number of operations gets reduced as the input size increases. Algorithms are found in binary tree or binary search functions. This involves the search of a given value in an array by splitting the array into two and starting searching in one split. This ensures the operation is not done on every element of the data.*

```python
#Prob_4 Take Suitable example and write function that runs in logarithic time. plot graph as well similar to above example.
import matplotlib.pyplot as plt
import math

def binary_search(arr, x):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] > x:
            high = mid - 1
        else:
            low = mid + 1
    return -1

arr = [10, 2, 3, 41, 22, 44]
x = 3
binary_search(arr, x)

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    binary_search(data, x)
    time_values.append(math.log2(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Binary Search Time Complexity")
plt.show()
```

#*Log-Linear Time Complexity O(n log n) : log-Â­linear running time*


* When we call a Logarithmic Time Algorithm inside a loop, it would
result into a Log-Linear Time Complexity program.




```python
#Prob_5 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example.
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result


arr = [38, 27, 43, 3, 9, 82, 10]
sorted_arr = merge_sort(arr)
print("Sorted array:", sorted_arr)

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    merge_sort(data)
    time_values.append(n*math.log2(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time ")
plt.title("O(nlog n) Binary Search Time Complexity")
plt.show()
```

### Polynomial Time Complexity O(n^c) : polynomial running time (c is a constant)

* When the computation time increases as function of N raised to some power, N being the size of input. Such a code has Polynomial Time complexity.




```python
# Prob_6 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example
def sum_arr(a):
  sum = 0
  for i in a:
    sum = sum + i
  return sum


arr = [10,20,30]
sum_arr(arr)

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    sum_arr(data)
    time_values.append(n)

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time")
plt.title("O(n) Linear Time Complexity")
plt.show()
```

#Exponential Time Complexity O(c^n) : exponential running time (c is a constant being raised to a power based on size of input)

* When the computation time of a code increases as function of X^N, N being the size of input. Such a code has Polynomial Time Complexity.




```python
# Prob_7 Take Suitable example and write function that runs in log-linear time. plot the graph as well, similar to above example

def log_linear(n):
    return 2**n


n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    log_linear(data)
    time_values.append(math.log2(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Quick sort Time Complexity")
plt.show()
```

```python
#Prob_8 plot graph similar to above figure
import math

def calculate_complexities(n):
  if n == 0:
    return 0, 0, 0, 0
  fn = n
  gn = math.log2(n)
  hn = n**2
  kn = 2**n
  return fn, gn, hn, kn

n_values = list(range(1, 11))
linear = []
log = []
polynomial= []
exponential= []

for n in n_values:
    fn, gn, hn, kn = calculate_complexities(n)
    linear.append(fn)
    log.append(gn)
    polynomial.append(hn)
    exponential.append(kn)

plt.figure(figsize=(10, 6))
plt.plot(n_values, linear, label='O(n)')
plt.plot(n_values, log, label='O(log n)')
plt.plot(n_values, polynomial, label='O(n^2)')
plt.plot(n_values, exponential, label='O(2^n)')

plt.xlabel('n ')
plt.ylabel('Time ')
plt.title('Comparison')
plt.legend()
plt.grid(True)
plt.ylim(bottom=0)
plt.show()
```

```python
# Write a function to perform quick sort and show that it has Logarithmic Time Complexity O(log n)
def quick_sort(arr):
    if len(arr) <= 1:
        return arr

    pivot = arr[-1]
    left, right = [], []

    for x in arr[:-1]:
        if x <= pivot:
            left.append(x)
        else:
            right.append(x)

    return quick_sort(left) + [pivot] + quick_sort(right)



numbers = [10, 7, 8, 9, 1, 5]
print("Sorted array:", quick_sort(numbers))

n_values = list(range(1, 101))
time_values = []

for n in n_values:
    data = list(range(n))
    quick_sort(data)
    time_values.append(math.log2(n))

plt.figure()
plt.plot(n_values, time_values)
plt.xlabel("Input size n")
plt.ylabel("Time (log n)")
plt.title("O(log n) Quick sort Time Complexity")
plt.show()
```
